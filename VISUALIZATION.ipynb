```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[100,150,200]
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title("house prices")
plt.xlabel("house number")
plt.ylabel("price")
plt.xlim(1,3)
plt.ylim(100,200);        #semicolon at the end of cell removes the output expression in the output box
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y,color='red',marker='o', linestyle="--")           #u can also give * as input in marker
```
